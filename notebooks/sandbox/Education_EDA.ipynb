```python
!pip install matplotlib
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', 50)
```

```python
edu = pd.read_csv('../../data/interim/EDUCATION.csv')
venues  = pd.read_csv('../../data/interim/CRYTO_VENUES_USA.csv')
venue_by_county = pd.read_csv('../../data/interim/VENUES_PER_COUNTY.csv')
```

```python
edu['Percent of adults with less than a high school diploma, 2015-19'].plot(kind='hist')

x = edu['Percent of adults with less than a high school diploma, 2015-19']
plt.subplot(2,2,1)
plt.title('less than a high school diploma')
# plt.xlabel('')
plt.hist(x)

x = edu['Percent of adults with a high school diploma only, 2015-19']
plt.subplot(2,2,2)
plt.title('high school diploma only')
# plt.xlabel('')
plt.hist(x)

x = edu['Percent of adults completing some college or associate\'s degree, 2015-19']
plt.subplot(2,2,3)
plt.title('some college or associate\'s degree')
# plt.xlabel('')
plt.hist(x)

x = edu['Percent of adults with a bachelor\'s degree or higher, 2015-19']
plt.subplot(2,2,4)
plt.title('bachelor\'s degree or higher')
# plt.xlabel('')
plt.hist(x)

plt.suptitle(' % of Adults Highest Level of Education 2015-2019', fontsize=20)
plt.tight_layout()
```

```python
edu_venue = edu.merge(venue_by_county, on=['state','county'], how='left')
edu_venue
```

```python
x = edu['Percent of adults with less than a high school diploma, 2015-19']
y= edu_venue['num_venues']
plt.subplot(2,2,1)
plt.title('less than a high school diploma')
# plt.xlabel('')
plt.scatter(x,y)

x = edu['Percent of adults with a high school diploma only, 2015-19']
y= edu_venue['num_venues']
plt.subplot(2,2,2)
plt.title('high school diploma only')
# plt.xlabel('')
plt.scatter(x,y)

x = edu['Percent of adults completing some college or associate\'s degree, 2015-19']
y= edu_venue['num_venues']
plt.subplot(2,2,3)
plt.title('some college or associate\'s degree')
# plt.xlabel('')
plt.scatter(x,y)

x = edu['Percent of adults with a bachelor\'s degree or higher, 2015-19']
y= edu_venue['num_venues']
plt.subplot(2,2,4)
plt.title('bachelor\'s degree or higher')
# plt.xlabel('')
plt.scatter(x,y)

plt.suptitle(' % of Adults Highest Level of Education 2015-2019', fontsize=20)
plt.tight_layout()
```
